- Record point value for winning player
- Display the score
- Continue playing until someone wins

```python
player1 = input("Enter Player1 name here:")
player2 = input("Enter Player2 name here:")

print("Player1: " + player1)
print("Player2: " + player2)
```

```python
import random

serve_first = random.randint(1,2)

if (serve_first == 1):
    print(player1 + " serves first!")
else:
    print(player2 + " serves first!")
```

```python
match_over = False

player1_point_score = 0
player1_game_score = 0
player1_set_score = 0

player2_point_score = 0
player2_game_score = 0
player2_set_score = 0

tiebreak = False

player1_previous_sets = [0, 0, 0]
player2_previous_sets = [0, 0, 0]
set_number = 1

while match_over != True:
    point_winner = random.randint(1,2)
    if tiebreak != True:
        if point_winner == 1:
            if (player1_point_score == 0) | (player1_point_score == 15):
                player1_point_score += 15
            elif (player1_point_score == 30):
                player1_point_score += 10
            elif (player1_point_score == 40):
                if (player2_point_score == 40):
                    player1_point_score += 20
                elif (player2_point_score == 60):
                    player2_point_score -= 20
                elif (player2_point_score == 0) & (player2_point_score == 15) & (player2_point_score == 30):
                    player1_point_score = 0
                    player1_game_score += 1
                    player2_point_score = 0
            elif (player1_point_score == 60):
                player1_point_score = 0
                player1_game_score += 1
                player2_point_score = 0
                
        if point_winner == 2:
            if (player2_point_score == 0) | (player2_point_score == 15):
                player2_point_score += 15
            elif (player2_point_score == 30):
                player2_point_score += 10
            elif (player2_point_score == 40):
                if (player1_point_score == 40):
                    player2_point_score += 20
                elif (player1_point_score == 60):
                    player1_point_score -= 20
                elif (player1_point_score == 0) & (player1_point_score == 15) & (player1_point_score == 30):
                    player2_point_score = 0
                    player2_game_score += 1
                    player1_point_score = 0
            elif (player2_point_score == 60):
                player2_point_score = 0
                player2_game_score += 1
                player1_point_score = 0
                
        print(player1 + "\t: " + str(player1_point_score) + " Game:" + str(player1_game_score) + " Set:" + str(player1_set_score))
        print(player2 + "\t: " + str(player2_point_score) + " Game:" + str(player2_game_score) + " Set:" + str(player2_set_score))
        print()
                
        if player1_game_score == 6:
            if player2_game_score < 5:
                player1_previous_sets[set_number - 1] = 6
                player2_previous_sets[set_number - 1] = player2_game_score
                player1_game_score = 0
                player1_set_score += 1
                player2_game_score = 0
                set_number += 1
            if player2_game_score == 6:
                tiebreak = True
                
        if (player1_game_score == 7) & (player2_game_score == 5):
            player1_previous_sets[set_number - 1] = 7
            player2_previous_sets[set_number - 1] = 5
            player1_game_score = 0
            player1_set_score += 1
            player2_game_score = 0
            set_number += 1
            
        if player2_game_score == 6:
            if player1_game_score < 5:
                player2_previous_sets[set_number - 1] = 6
                player1_previous_sets[set_number - 1] = player1_game_score
                player2_game_score = 0
                player2_set_score += 1
                player1_game_score = 0
                set_number += 1
            if player1_game_score == 6:
                tiebreak = True
                
        if (player2_game_score == 7) & (player1_game_score == 5):
            player2_previous_sets[set_number - 1] = 7
            player1_previous_sets[set_number - 1] = 5
            player2_game_score = 0
            player2_set_score += 1
            player1_game_score = 0
            set_number += 1
            
    else: #tiebreak 
        if point_winner == 1:
            player1_point_score += 1
        elif point_winner == 2:
            player2_point_score += 1
            
        print(player1 + "\t: " + str(player1_point_score) + " Game:" + str(player1_game_score) + " Set:" + str(player1_set_score))
        print(player2 + "\t: " + str(player2_point_score) + " Game:" + str(player2_game_score) + " Set:" + str(player2_set_score))
        print()
            
        if (player1_point_score >= 7) & (player1_point_score >= player2_point_score + 2):
            player1_previous_sets[set_number - 1] = 7
            player2_previous_sets[set_number - 1] = 6
            player1_point_score = 0
            player1_game_score = 0
            player1_set_score += 1
            player2_point_score = 0
            player2_game_score = 0
            set_number += 1
            tiebreak = False
            
        if (player2_point_score >= 7) & (player2_point_score >= player1_point_score + 2):
            player2_previous_sets[set_number - 1] = 7
            player1_previous_sets[set_number - 1] = 6
            player2_point_score = 0
            player2_game_score = 0
            player2_set_score += 1
            player1_point_score = 0
            player1_game_score = 0
            set_number += 1
            tiebreak = False
    
    if (player1_set_score == 2) | (player2_set_score == 2):
        match_over = True
        
if (player1_set_score == 2):
    winner = player1
else:
    winner = player2

print("The winner is " + winner + "!")
print(player1 + "\t: " + str(player1_previous_sets[0]) + " " + str(player1_previous_sets[1]) + " " + str(player1_previous_sets[2]))
print(player2 + "\t: " + str(player2_previous_sets[0]) + " " + str(player2_previous_sets[1]) + " " + str(player2_previous_sets[2]))
```

```python
#This is a function that converts regular point-based scoring into traditional tennis scores
def point_to_tennis_score(player1_point_score, player2_point_score):
    if player1_point_score == 0:
        tennis_score = "0"
    elif player1_point_score == 1:
        tennis_score = "15"
    elif player1_point_score == 2:
        tennis_score = "30"
    elif (player1_point_score == 3):
        tennis_score = "40"
    elif (player1_point_score > 3):
        if (player1_point_score <= player2_point_score):
            tennis_score = "40"
        if (player1_point_score > player2_point_score):
            tennis_score = "Ad"
    return tennis_score

#Function that switches who serves next
def switch_server(server):
    if server == 1:
        server = 2
    elif server == 2:
        server = 1
    return server
    
def check_game_point(player1, player2):
    return_value = 0
    
    if ((player1_point_score == 3) & (player2_point_score <= 2)) | ((player1_point_score > 3) & (player1_point_score - player2_point_score == 1)):
        return_value = 1
    elif ((player2_point_score == 3) & (player1_point_score <= 2)) | ((player2_point_score > 3) & (player2_point_score - player1_point_score == 1)):
        return_value = 2
    
    return return_value
        
def check_set_point(player1, player2, is_game_point):
    return_value = 0
    
    if (is_game_point == 1) & ((player1_game_score == 5) & (player2_game_score <= 4) | (player1_game_score == 6) & (player2_game_score == 5)):
        return_value = 1
    elif (is_game_point == 2) & ((player2_game_score == 5) & (player1_game_score <= 4) | (player2_game_score == 6) & (player1_game_score == 5)):
        return_value = 2
        
    return return_value
        
def check_tiebreak_point(player1, player2):
    return_value = 0
    
    if (player1_point_score == 6) & (player2_point_score <= 5) | (player1_point_score > 6) & (player1_point_score - player2_point_score == 1):
        return_value = 1
    elif (player2_point_score == 6) & (player1_point_score <= 5) | (player2_point_score > 6) & (player2_point_score - player1_point_score == 1):
        return_value = 2
        
    return return_value

def check_match_point(player1, player2, is_set_point):
    return_value = 0
    
    if (is_set_point == 1) & ((player1_set_score == 1) & (player2_set_score <= 1)):
        return_value = 1
    elif (is_set_point == 2) & ((player2_set_score == 1) & (player1_set_score <= 1)):
        return_value = 2
        
    return return_value
        
def check_break_point(player1, player2, is_game_point):
    return_value = 0
    
    if (is_game_point == 1) & (server == 2):
        return_value = 1
    elif (is_game_point == 2) & (server == 1):
        return_value = 2
        
    return return_value

#this function checks to see if a player has saved any break points
def check_saved_break_point(player1, player2, is_break_point, point_winner):
    #is_break_point is the return value of check_break_point() from the previous while loop
    #point_winner indicates the current point winner from the current loop
    if (is_break_point == 1) & (point_winner == 2):
        print(player2 + " has saved a break point!\n")
    elif (is_break_point == 2) & (point_winner == 1):
        print(player1 + " has saved a break point!\n")

match_over = False

#Variables for scorekeeping
player1_point_score = 0
player1_game_score = 0
player1_set_score = 0

player2_point_score = 0
player2_game_score = 0
player2_set_score = 0


tiebreak = False

#Recording the game values of previous sets and of who won the set
player1_previous_sets = [0, 0, 0]
player2_previous_sets = [0, 0, 0]
set_number = 1

game_winner = 0
set_winner = 0

if (server == 1):
    player1s = player1 + "*"
    player2s = player2
else:
    player1s = player1
    player2s = player2 + "*"

#This max() function determines the longer of the two players' names
#and then the function inputs it into the name slot
x = max(len(player1), len(player2)) + 2
y = "{:<" + str(x) + "}:"
print(y.format(player1s) + " 0" + " Game: " + str(player1_game_score) + " Set: " + str(player1_set_score))
print(y.format(player2s) + " 0" + " Game: " + str(player2_game_score) + " Set: " + str(player2_set_score))
print()

player1_first_serve_percentage = 0.68 
player1_first_serve_win_rate = 0.72
player1_second_serve_percentage = 0.97
player1_second_serve_win_rate = 0.57

player2_first_serve_percentage = 0.62 
player2_first_serve_win_rate = 0.77
player2_second_serve_percentage = 0.98
player2_second_serve_win_rate = 0.57

#Beginning of code, this section is active when the match is still going on
while match_over != True:
    #Randomly selects the winner of that certain point
#     if random.random() <= 0.6:
#         point_winner = server
#     else:
#         #this means that 40% of the time the returner will win the point
#         if server == 1:
#             point_winner = 2
#         else:
#             point_winner = 1 

    if server == 1:
        if random.random() <= player1_first_serve_percentage:
            if random.random() <= player1_first_serve_win_rate:
                point_winner = 1
            else:
                point_winner = 2
        else:
            if random.random() <= player1_second_serve_percentage:
                if random.random() <= player1_second_serve_win_rate:
                    point_winner = 1
                else:
                    point_winner = 2
            else:
                point_winner = 2
                
    elif server == 2:
        if random.random() <= player2_first_serve_percentage:
            if random.random() <= player2_first_serve_win_rate:
                point_winner = 2
            else:
                point_winner = 1
        else:
            if random.random() <= player2_second_serve_percentage:
                if random.random() <= player2_second_serve_win_rate:
                    point_winner = 2
                else:
                    point_winner = 1
            else:
                point_winner = 1
    
    #When the game is not a tiebreak game
    if tiebreak != True:
        
        
        #When the first player wins the point
        if point_winner == 1:
            player1_point_score += 1
        #When the first player wins the game
        if ((player1_point_score == 4) & (player2_point_score <= 2)) | \
        ((player1_point_score >= 3) & (player2_point_score >= 3) & \
         (player1_point_score - player2_point_score == 2)):
            player1_point_score = 0
            player1_game_score += 1
            player2_point_score = 0
            game_winner = 1
            server = switch_server(server)
            
        #When the second player wins the point
        if point_winner == 2:
            player2_point_score += 1
        #When the second player wins the game
        if ((player2_point_score == 4) & (player1_point_score <= 2)) | \
        ((player2_point_score >= 3) & (player1_point_score >= 3) & \
         (player2_point_score - player1_point_score == 2)):
            player2_point_score = 0
            player2_game_score += 1
            player1_point_score = 0
            game_winner = 2
            server = switch_server(server)
        
            
        #What happens when the first player reaches 6 games
        if player1_game_score == 6:
            #What happens when the first player wins the set
            if player2_game_score < 5:
                player1_previous_sets[set_number - 1] = 6
                player2_previous_sets[set_number - 1] = player2_game_score
                player1_set_score += 1
                set_number += 1
                set_winner = 1
                
            #What happens when the set reaches a tiebreak
            if player2_game_score == 6:
                tiebreak = True
                
        #When the first player wins the set 7-5
        if (player1_game_score == 7) & (player2_game_score == 5):
            player1_previous_sets[set_number - 1] = 7
            player2_previous_sets[set_number - 1] = 5
            player1_set_score += 1
            set_number += 1
            set_winner = 1
            
        #When the second player reaches 6 games
        if player2_game_score == 6:
            #when the second player wins the set
            if player1_game_score < 5:
                player2_previous_sets[set_number - 1] = 6
                player1_previous_sets[set_number - 1] = player1_game_score
                player2_set_score += 1
                set_number += 1
                set_winner = 2
            #When the set goes into a tiebreak
            if player1_game_score == 6:
                tiebreak = True
                
        #When the second player wins the set 7-5
        if (player2_game_score == 7) & (player1_game_score == 5):
            player2_previous_sets[set_number - 1] = 7
            player1_previous_sets[set_number - 1] = 5
            player2_set_score += 1
            set_number += 1
            set_winner = 2
            
        #When the first player serves, add an asterik
        if (server == 1):
            player1s = player1 + "*"
            player2s = player2
        else:
            player1s = player1
            player2s = player2 + "*"
            
        if (game_winner != 0):
            if (set_winner == 0):
                if game_winner == 1:
                    print(player1 + " won the game!\n")
                elif game_winner == 2:
                    print(player2 + " won the game!\n")
            elif (set_winner != 0):   
                print(y.format(player1s) + " Game: " + str(player1_game_score) + " Set: " + str(player1_set_score))
                print(y.format(player2s) + " Game: " + str(player2_game_score) + " Set: " + str(player2_set_score))
                print()
                if set_winner == 1:
                    print(player1 + " won the game and the set!\n")
                elif (set_winner == 2):
                    print(player2 + " won the game and the set!\n") 
                player1_game_score = 0
                player2_game_score = 0
                set_winner = 0
                if (player1_set_score == 2) | (player2_set_score == 2):
                    match_over = True
            game_winner = 0
                    
        if match_over != True:
            print(y.format(player1s) + " " + point_to_tennis_score(player1_point_score, player2_point_score) + 
                  " Game: " + str(player1_game_score) + " Set: " + str(player1_set_score))
            print(y.format(player2s) + " " + point_to_tennis_score(player2_point_score, player1_point_score) +
                  " Game: " + str(player2_game_score) + " Set: " + str(player2_set_score))
            print()
                
        #check_saved_break_point() is placed before is_break_point because this way, 
        #check_saved_break_point() checks if there is a break point in the previous loop,
        #while if we had placed check_saved_break_point() after check_break_point(),
        #the function 
#         check_saved_break_point(player1, player2, is_break_point, point_winner)
            is_game_point = check_game_point(player1, player2)
            is_set_point = check_set_point(player1, player2, is_game_point)
            is_break_point = check_break_point(player1, player2, is_game_point)
            is_match_point = check_match_point(player1, player2, is_set_point)
    #         if is_game_point == 1:
    #             if is_set_point == 0:
    #                 print(player1 + " has a game point!\n")
    #             elif is_set_point == 1:
    #                 print(player1 + " has a set point!\n")
    #         elif is_game_point == 2:
    #             if is_set_point == 0:
    #                 print(player2 + " has a game point!\n")
    #             elif is_set_point == 2:
    #                 print(player2 + " has a set point!\n")

            #check is_set_point() first, and then check is_game_point
            #because is_set_point is a subset of is_game_point
            if is_match_point == 1:
                print(player1 + " has a match point!\n")
            elif is_match_point == 2:
                print(player2 + " has a match point!\n")
            elif is_match_point == 0:
                if is_set_point == 1:
                    print(player1 + " has a set point!\n")
                elif is_set_point == 2:
                    print(player2 + " has a set point!\n")
                elif is_set_point == 0:
                    if is_break_point == 1:
                        print(player1 + " has a break point!\n")
                    elif is_break_point == 2:
                        print(player2 + " has a break point!\n")
                    elif is_break_point == 0:
                        if is_game_point == 1:
                            print(player1 + " has a game point!\n")
                        elif is_game_point == 2:
                            print(player2 + " has a game point!\n")
        
            
        
    else: #tiebreak 
        if point_winner == 1:
            player1_point_score += 1
            #print_set_point(player1, player2)
        elif point_winner == 2:
            player2_point_score += 1
            #print_set_point(player2, player1)
            
        if (((player1_point_score + player2_point_score) % 2) == 1):
            server = switch_server(server)
            
        #is_set_point = check_tiebreak_point(player1, player2)
        #check_match_point(player1, player2, is_set_point)
        
        #When the first player wins the tiebreak    
        if (player1_point_score >= 7) & (player1_point_score >= player2_point_score + 2):
            player1_previous_sets[set_number - 1] = 7
            player2_previous_sets[set_number - 1] = 6
            player1_point_score = 0
            player1_game_score += 1
            player1_set_score += 1
            player2_point_score = 0
            set_number += 1
            set_winner = 1
            tiebreak = False
            
        #When the second player wins the tiebreak
        if (player2_point_score >= 7) & (player2_point_score >= player1_point_score + 2):
            player2_previous_sets[set_number - 1] = 7
            player1_previous_sets[set_number - 1] = 6
            player2_point_score = 0
            player2_game_score += 1
            player2_set_score += 1
            player1_point_score = 0
            set_number += 1
            set_winner = 2
            tiebreak = False
            
        if (server == 1):
            player1s = player1 + "*"
            player2s = player2
        else:
            player1s = player1
            player2s = player2 + "*"

        #print everything at the bottom,
        #this makes things easier to read and it is more organized
        #if there are redundant parts of code in different if-else statements,
        #you can bring them outside and after the if-else statements
        #these pieces of code are to be run anyway, so putting them at the end simplifies things
        if (set_winner != 0):    
                print(y.format(player1s) + " Game: " + str(player1_game_score) + " Set: " + str(player1_set_score))
                print(y.format(player2s) + " Game: " + str(player2_game_score) + " Set: " + str(player2_set_score))
                print()
                if set_winner == 1:
                    print(player1 + " won the set!\n")
                elif (set_winner == 2):
                    print(player2 + " won the set!\n")
                #this code is to reset the set_winner, 
                #so that the program does not print the wrong set winner
                player1_game_score = 0
                player2_game_score = 0
                set_winner = 0
                if (player1_set_score == 2) | (player2_set_score == 2):
                    match_over = True
        #printing the point score for the tiebreak,
        #and if the tiebreak ends, 
        #print the first score for the next game!!!!!
        if match_over != True:
            print(y.format(player1s) + str(player1_point_score) + " Game: " + str(player1_game_score) + " Set: " + str(player1_set_score))
            print(y.format(player2s) + str(player2_point_score) + " Game: " + str(player2_game_score) + " Set: " + str(player2_set_score))
            print()
            
            is_tiebreak_point = check_tiebreak_point(player1, player2)
            is_match_point = check_match_point(player1, player2, is_tiebreak_point)
            
            if is_match_point == 1:
                print(player1 + " has a match point!\n")
            elif is_match_point == 2:
                print(player2 + " has a match point!\n")
            elif is_match_point == 0:
                if is_tiebreak_point == 1:
                    print(player1 + " has a set point!\n")
                elif is_tiebreak_point == 2:
                    print(player2 + " has a set point!\n")
        
if (player1_set_score == 2):
    match_winner = player1
else:
    match_winner = player2
    
print(match_winner + " won the match!\n")
print(y.format(player1) + " " + str(player1_previous_sets[0]) + " " + str(player1_previous_sets[1]) + " " + str(player1_previous_sets[2]))
print(y.format(player2) + " " + str(player2_previous_sets[0]) + " " + str(player2_previous_sets[1]) + " " + str(player2_previous_sets[2]))
```

```python
import pandas as pd

class Match_Manager:
        
    def __init__(self, player1, player2):
        self.match_over = False
        self.tiebreak = False
        #Recording the game values of previous sets and of who won the set
        self.set_number = 0
        self.point_winner = 0
        self.game_winner = False
        self.set_winner = False
        self.match_winner = None
        
        self.players = []

        self.server = None
        
        #appends the players directly from spreadsheet
        self.players.append(player1)
        player1.index = 0
        self.players.append(player2)
        player2.index = 1
        
        column_names = ["name", "point_score", "game_score", "set_score", "break_point", "set_point",
                        "match_point", "break_point_saved", "set_point_saved", "match_point_saved"]
        self.df_scores = pd.DataFrame(columns = column_names)
        
    def reset(self):
        self.match_over = False
        self.tiebreak = False
        #Recording the game values of previous sets and of who won the set
        self.set_number = 0
        self.point_winner = 0
        self.game_winner = False
        self.set_winner = False
        
#         self.players = []

        self.server = None

    #This is a function that converts regular point-based scoring into traditional tennis scores
    def point_to_tennis_score(self, player_index, players):
        if players[player_index].point_score == 0:
            tennis_score = "0"
        elif players[player_index].point_score == 1:
            tennis_score = "15"
        elif players[player_index].point_score == 2:
            tennis_score = "30"
        elif (players[player_index].point_score == 3):
            tennis_score = "40"
        elif (players[player_index].point_score > 3):
            if (players[player_index].point_score <= players[self.opponent(player_index)].point_score):
                tennis_score = "40"
            if (players[player_index].point_score > players[self.opponent(player_index)].point_score):
                tennis_score = "Ad"
        return tennis_score

    #Function that switches who serves next
    def switch_server(self, server):
        return self.opponent(server)

    def check_game_point(self, players):
        return_value = None

        for i in range(0,len(players)):
            if ((players[i].point_score == 3) & (players[self.opponent(i)].point_score <= 2)) | \
                ((players[i].point_score > 3) & (players[i].point_score - players[self.opponent(i)].point_score == 1)):
                return_value = i

        return return_value

    def check_set_point(self, players, game_point_player_index):
        return_value = None

        if (game_point_player_index != None):
            if ((players[game_point_player_index].game_score == 5) & \
                (players[self.opponent(game_point_player_index)].game_score <= 4) | 
                (players[game_point_player_index].game_score == 6) & \
                (players[self.opponent(game_point_player_index)].game_score == 5)):
                return_value = game_point_player_index

        return return_value

    def check_tiebreak_point(self, players):
        return_value = None

        for i in range(0,len(players)):
            if ((players[i].point_score == 6) & \
                (players[self.opponent(i)].point_score <= 5)) | \
                ((players[i].point_score > 6) & 
                 (players[i].point_score - players[self.opponent(i)].point_score == 1)):
                return_value = i

        return return_value

    def check_match_point(self, players, set_point_player_index):
        return_value = None

        if (set_point_player_index != None):
            if ((players[set_point_player_index].set_score == 1) & 
                (players[self.opponent(set_point_player_index)].set_score <= 1)):
                return_value = set_point_player_index

        return return_value

    def check_break_point(self, players, game_point_player_index):
        return_value = None

        if (game_point_player_index != None): 
            if (self.server != game_point_player_index):
                players[game_point_player_index].break_points_count += 1
                return_value = game_point_player_index

        return return_value

    #this function checks to see if a player has saved any break points
    def check_saved_break_point(self, players, server, point_winner_index):
        return_value = None

        if point_winner_index != None:
            #opponent(server) is a function, and its return value is an index, just like check_game_point(players),
            #so calling opponent(server) makes the types match
            if (point_winner_index == server):
                if (players[self.opponent(server)].point_score >= 3) & \
                    (players[self.opponent(server)].point_score >= players[server].point_score):
                    players[point_winner_index].break_points_saved_count += 1
                    return_value = point_winner_index

        return return_value

    def check_saved_set_point(self, players, point_winner_index):
        return_value = None

        if point_winner_index != None:
            if ((players[self.opponent(point_winner_index)].game_score >= 5) & (players[self.opponent(point_winner_index)].game_score < 7)) & \
                (players[self.opponent(point_winner_index)].game_score > players[point_winner_index].game_score):
                if (players[self.opponent(point_winner_index)].point_score >= 3) & \
                    (players[self.opponent(point_winner_index)].point_score >= players[point_winner_index].point_score):
                    return_value = point_winner_index

        return return_value

    def check_saved_match_point(self, players, point_winner_index):
        return_value = None

        if point_winner_index != None:
            if (players[self.opponent(point_winner_index)].set_score == 1) & (players[point_winner_index].set_score <= 1):
                if self.check_saved_set_point(players, point_winner_index) != None:
                    return_value = point_winner_index

        return return_value

    def check_saved_match_point_in_tiebreak(self, players, point_winner_index):
        return_value = None

        if point_winner_index != None:
            if (players[self.opponent(point_winner_index)].set_score == 1) & (players[point_winner_index].set_score <= 1):
                if (players[self.opponent(point_winner_index)].point_score >= 6) & \
                    (players[self.opponent(point_winner_index)].point_score >= players[point_winner_index].point_score):
                    return_value = point_winner_index

        return return_value

    def opponent(self, player_index):
        if player_index == 0:
            return 1
        else:
            return 0
        
    def match_simulator(self, print_scores = True, print_stats = True):

        for player in self.players:
            if player.name == "Bye":
                self.match_winner = self.players[self.opponent(player.index)]
                return None
        
        
        self.reset()
        for player in self.players:
            player.reset()

        #Determines who serves first via random integer function
        self.server = random.randint(0,1)

        #Prints whoever is serving first
        if print_scores:
            print(self.players[self.server].name + " serves first!")
            print()

        #This adds an asterik to the name of the server 
        #and removes the star from the opponent's name
        self.players[self.server].serve_name = self.players[self.server].name + "*"
        self.players[self.opponent(self.server)].serve_name = self.players[self.opponent(self.server)].name

        #This max() function determines the longer of the two players' names
        #and then the function inputs it into the name slot
        name_length = []

        for player in self.players:
            name_length.append(len(player.name))

        #This piece looks through the name_length[] list
        #and finds the longest name's length and adds 2 to the end
        #to accomodate for the asterik and is easier to read
        x = max(name_length) + 2
        y = "{:<" + str(x) + "}:"

        if print_scores:
            for player in self.players:
                print(y.format(player.serve_name) + " 0" + " Game: " + str(player.game_score) + " Set: " + str(player.set_score))
            print()

        #Beginning of code, this section is active when the match is still going on
        while self.match_over != True:

            self.players[self.server].first_serve_count += 1
            if random.random() <= self.players[self.server].first_serve_percentage:
                self.players[self.server].first_serve_in_count += 1
                if random.random() <= self.players[self.server].first_serve_win_rate:
                    self.players[self.server].first_serve_win_count += 1
                    self.point_winner = self.server
                else:
                    self.point_winner = self.opponent(self.server)
            else:
                #First serve fault
                self.players[self.server].second_serve_count += 1
                if random.random() <= self.players[self.server].second_serve_percentage:
                    self.players[self.server].second_serve_in_count += 1
                    if random.random() <= self.players[self.server].second_serve_win_rate:
                        self.players[self.server].second_serve_win_count += 1
                        self.point_winner = self.server
                    else:
                        self.point_winner = self.opponent(self.server)
                else:
                    #Double fault
                    self.players[self.server].double_fault_count += 1
                    self.point_winner = self.opponent(self.server)

            self.players[self.point_winner].points_won_count += 1

            #When the game is not a tiebreak game
            if self.tiebreak != True:


                #When a player wins the point
                self.players[self.point_winner].point_score += 1
                #When the first player wins the game
                if ((self.players[self.point_winner].point_score == 4) & (self.players[self.opponent(self.point_winner)].point_score <= 2)) | \
                ((self.players[self.point_winner].point_score >= 3) & (self.players[self.opponent(self.point_winner)].point_score >= 3) & \
                 (self.players[self.point_winner].point_score - self.players[self.opponent(self.point_winner)].point_score == 2)):
                    self.players[self.point_winner].point_score = 0
                    self.players[self.point_winner].game_score += 1
                    self.players[self.opponent(self.point_winner)].point_score = 0
                    self.game_winner = True
                    self.server = self.switch_server(self.server)


                #What happens when the first player reaches 6 games
                if self.players[self.point_winner].game_score == 6:
                    #What happens when the first player wins the set
                    if self.players[self.opponent(self.point_winner)].game_score < 5:
                        self.players[self.point_winner].previous_sets[self.set_number] = 6
                        self.players[self.opponent(self.point_winner)].previous_sets[self.set_number] = self.players[self.opponent(self.point_winner)].game_score
                        self.players[self.point_winner].set_score += 1
                        self.set_number += 1
                        self.set_winner = True
                    #What happens when the set reaches a tiebreak
                    if self.players[self.opponent(self.point_winner)].game_score == 6:
                        self.tiebreak = True

                #When the first player wins the set 7-5
                if (self.players[self.point_winner].game_score == 7) & (self.players[self.opponent(self.point_winner)].game_score == 5):
                    self.players[self.point_winner].previous_sets[self.set_number] = 7
                    self.players[self.opponent(self.point_winner)].previous_sets[self.set_number] = 5
                    self.players[self.point_winner].set_score += 1
                    self.set_number += 1
                    self.set_winner = True

                #When the first player serves, add an asterik
                self.players[self.server].serve_name = self.players[self.server].name + "*"
                self.players[self.opponent(self.server)].serve_name = self.players[self.opponent(self.server)].name

                if (self.game_winner == True):
                    if (self.set_winner != True):
                        if print_scores:
                            print(self.players[self.point_winner].name + " won the game!\n")
                    else:
                        for player in self.players:
                            if print_scores:
                                print(y.format(player.serve_name) + " Game: " + str(player.game_score) + " Set: " + str(player.set_score))
                        if print_scores:
                            print()
                            print(self.players[self.point_winner].name + " won the game and the set!\n") 
                        for player in self.players:
                            player.game_score = 0
                        self.set_winner = False
                        for player in self.players:
                            if player.set_score == 2:
                                self.match_over = True
                    self.game_winner = False

                if self.match_over != True:
                    for i in range(0,len(self.players)):
                        if print_scores:
                            print(y.format(self.players[i].serve_name) + " " + self.point_to_tennis_score(i, self.players) + 
                              " Game: " + str(self.players[i].game_score) + " Set: " + str(self.players[i].set_score))
                        dict = {"name" : self.players[i].serve_name,
                               "point_score" : self.point_to_tennis_score(i, self.players),
                               "game_score" : (self.players[i].game_score),
                               "set_score" : (self.players[i].set_score)}
                        self.df_scores = self.df_scores.append(dict, ignore_index = True)
                    if print_scores:
                        print()
                    
                    

                #check_saved_break_point() is placed before is_break_point because this way, 
                #check_saved_break_point() checks if there is a break point in the previous loop,
                #while if we had placed check_saved_break_point() after check_break_point(),
                #the function 
                    game_point_player_index = self.check_game_point(self.players)
                    set_point_player_index = self.check_set_point(self.players, game_point_player_index)
                    break_point_player_index = self.check_break_point(self.players, game_point_player_index)
                    match_point_player_index = self.check_match_point(self.players, set_point_player_index)
                    saved_break_point_index = self.check_saved_break_point(self.players, self.server, self.point_winner)
                    saved_set_point_index = self.check_saved_set_point(self.players, self.point_winner)
                    saved_match_point_index = self.check_saved_match_point(self.players, self.point_winner)
                    #check is_set_point() first, and then check is_game_point
                    #because is_set_point is a subset of is_game_point
                    if match_point_player_index != None:
                        if print_scores:
                            print(self.players[match_point_player_index].name + " has a match point!\n")
                    else:
                        if set_point_player_index != None:
                            if print_scores:
                                print(self.players[set_point_player_index].name + " has a set point!\n")
                        else:
                            if break_point_player_index != None:
                                if print_scores:
                                    print(self.players[break_point_player_index].name + " has a break point!\n")
                            else:
                                if game_point_player_index != None:
                                    if print_scores:
                                        print(self.players[game_point_player_index].name + " has a game point!\n")
                            #Don't put saved_break_point_index after break_point_player_index,
                            #as this messes up the else: into relating to saved_break_point_index
                    if saved_match_point_index != None:
                        if print_scores:
                            print(self.players[saved_match_point_index].name + " has saved a match point!\n")
                    else:
                        if saved_set_point_index != None:
                            if print_scores:
                                print(self.players[saved_set_point_index].name + " has saved a set point!\n")
                        else:
                            if saved_break_point_index != None:
                                if print_scores:
                                    print(self.players[saved_break_point_index].name + " has saved a break point!\n")


            else: #tiebreak 
                self.players[self.point_winner].point_score += 1

                if (((self.players[self.server].point_score + self.players[self.opponent(self.server)].point_score) % 2) == 1):
                    self.server = self.switch_server(self.server)

                #When any player wins the tiebreak    
                if ((self.players[self.point_winner].point_score >= 7) & 
                    (self.players[self.point_winner].point_score >= self.players[self.opponent(self.point_winner)].point_score + 2)):
                    self.players[self.point_winner].previous_sets[self.set_number] = 7
                    self.players[self.opponent(self.point_winner)].previous_sets[self.set_number] = 6
                    self.players[self.point_winner].point_score = 0
                    self.players[self.point_winner].game_score += 1
                    self.players[self.point_winner].set_score += 1
                    self.players[self.opponent(self.point_winner)].point_score = 0
                    self.set_number += 1
                    self.set_winner = True
                    self.tiebreak = False

                self.players[self.server].serve_name = self.players[self.server].name + "*"
                self.players[self.opponent(self.server)].serve_name = self.players[self.opponent(self.server)].name

                #print everything at the bottom,
                #this makes things easier to read and it is more organized
                #if there are redundant parts of code in different if-else statements,
                #you can bring them outside and after the if-else statements
                #these pieces of code are to be run anyway, so putting them at the end simplifies things
                if (self.set_winner == True):
                    for player in self.players:
                        if print_scores:
                            print(y.format(player.serve_name) + " Game: " + str(player.game_score) + " Set: " + str(player.set_score))
                    if print_scores:
                        print()
                        print(self.players[self.point_winner].name + " won the set!\n")
                    #this code is to reset the set_winner, 
                    #so that the program does not print the wrong set winner
                    for player in self.players:
                        player.game_score = 0
                    self.set_winner = False
                    for player in self.players:
                        if (player.set_score == 2):
                            self.match_over = True
                #printing the point score for the tiebreak,
                #and if the tiebreak ends, 
                #print the first score for the next game!!!!!
                if self.match_over != True:
                    for player in self.players:
                        if print_scores:
                            print(y.format(player.serve_name) + str(player.point_score) + " Game: " + str(player.game_score) + " Set: " + str(player.set_score))
                        dict = {"name" : player.serve_name,
                               "point_score" : str(player.point_score),
                               "game_score" : (player.game_score),
                               "set_score" : (player.set_score)}
                        self.df_scores = self.df_scores.append(dict, ignore_index = True)
                    if print_scores:
                        print()

                    tiebreak_point_player_index = self.check_tiebreak_point(self.players)
                    match_point_player_index = self.check_match_point(self.players, tiebreak_point_player_index)
                    saved_match_point_in_tiebreak_index = self.check_saved_match_point_in_tiebreak(self.players, self.point_winner)

                    if match_point_player_index != None:
                        if print_scores:
                            print(self.players[match_point_player_index].name + " has a match point!\n")
                    else:
                        if tiebreak_point_player_index != None:
                            if print_scores:
                                print(self.players[tiebreak_point_player_index].name + " has a set point!\n")
                    if saved_match_point_in_tiebreak_index != None:
                        if print_scores:
                            print(self.players[saved_match_point_in_tiebreak_index].name + " has saved a match point!\n")

        for player in self.players:        
            if (player.set_score == 2):
                self.match_winner = player
        if print_scores:
            print(self.match_winner.name + " won the match!\n")
        for player in self.players: 
            if print_scores:
                print(y.format(player.name) + " " + str(player.previous_sets[0]) + " " + str(player.previous_sets[1]) + " " + str(player.previous_sets[2]))

        for player in self.players:
            if print_stats:
                print("%s had a %d%% first serve rate." %(y.format(player.name),int((player.first_serve_in_count/player.first_serve_count) * 100)))
                print("%s had a %d%% second serve rate." %(y.format(player.name),int((player.second_serve_in_count/player.second_serve_count) * 100)))
                print("%s served %d double faults." %(y.format(player.name),player.double_fault_count))
                print("%s won %d%% of his first serve points." %(y.format(player.name), int((player.first_serve_win_count/player.first_serve_in_count) * 100)))
                print("%s won %d%% of his second serve points." %(y.format(player.name), int((player.second_serve_win_count/player.second_serve_in_count) * 100)))
                print("%s saved %d out of %d break points." %(y.format(player.name), player.break_points_saved_count, self.players[self.opponent(player.index)].break_points_count))
                print("%s won %d points in total." %(y.format(player.name), player.points_won_count))
                print()
```
